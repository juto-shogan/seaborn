```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
crash_df = sns.load_dataset('car_crashes')
tips_df = sns.load_dataset('tips')
```

Categorical Plots

bar Plots

```python
sns.barplot(x='sex', y='total_bill', data= tips_df, estimator=np.median)
```

```python
# tested if the above graph could be ploted in matplot and it can

plt.bar(tips_df['sex'], height=0.1
        * np.random.randn(len(tips_df['sex'])),
        color='blue',
        alpha=0.5)

plt.xticks()

plt.title('Bar Chart')
plt.title('Bar Chart')
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')
```

count plot


```python
sns.countplot(data= tips_df, x= 'sex')
```

box plot

```python
sns.boxplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex')# hue adds another category
```

violin plot

```python
plt.figure(figsize=(10,5))
sns.violinplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex', split=True)
plt.legend(loc=0)
```

strip plots

```python
plt.figure(figsize=(7, 10))
sns.stripplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex', dodge=True)# dodge is to seperate the categories

plt.legend(loc=1)
```

```python
#sns.violinplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex', split=True)
sns.swarmplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex')
```

palettes

```python
plt.figure(figsize=(5,10))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(data= tips_df, x= 'day', y= 'total_bill', hue='sex', palette= 'magma'  )


plt.legend(loc=0)
```
