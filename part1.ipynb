```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
sns.displot(crash_df['not_distracted'], kde= True, bins=25)#kde adds the line to the plot, bins are used to adjust the spread

plt.savefig('dist.png', dpi=80)
```

joint Plot

```python
plt.figure(figsize=(10,5))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hist')

plt.savefig('joint.png', dpi=80)
```

kde Plot

```python
sns.kdeplot(crash_df['alcohol'])

plt.savefig('kde.png', dpi=80)
```

pair Plots

```python
sns.pairplot(crash_df)

plt.savefig('pair.png', dpi=300)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette = 'Blues')

plt.savefig('pair_hue.png', dpi=300)
```

rug Plots

```python
sns.rugplot(tips_df['tip'])

plt.savefig('rug.png', dpi=80)
```
